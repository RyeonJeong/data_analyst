```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data = pd.read_csv("./data/ThoraricSurgery3.csv", header =None)
data.head(2)
```

```python
X = data.drop(16, axis=1)
y = data[16]
```

```python
model = Sequential()
model.add(Dense(30, input_dim=16, activation='relu'))
model.add(Dense(60, input_dim=16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# Dense가 여러개 쌓이면 딥러닝, 메모리 상승 시간 오래걸릴뿐
```

```python
%%time
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X, y, epochs=50, batch_size=16)
# batch_size : 데이터 캣수 사이즈 8 16 32 ... 
# epochs : 훈련 횟수
# batch_size가 너무 크면(과적합) 메모리를 너무 먹어서 컴퓨터 다운, 적절히 조정하기(가능한 크게)
# batch_size가 너무 낮으면 데이터 쪼개기 때문에 성능 떨어짐
# epochs 많이 넣으면 과적합, 적으면 성능 떨어짐
```
