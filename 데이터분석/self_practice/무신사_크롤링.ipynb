```python
# ì…€ë ˆë‹ˆì›€ìœ¼ë¡œ ê°€ì ¸ì˜¤ê¸°
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.set_window_size(1920, 1080) # ì›¹ ë¸Œë¼ìš°ì € í•´ìƒë„ ì¡°ì ˆ
driver.get("https://www.musinsa.com/ranking/best?period=now&age=ALL&mainCategory=&subCategory=&leafCategory=&price=&golf=false&kids=false&newProduct=false&exclusive=false&discount=false&soldOut=false&page=1&viewType=small&priceMin=&priceMax=")
```

```python
soup2 = bs(driver.page_source, 'lxml')
```

```python
soup2.select('ul.snap-article-list')
```

```python
product_list = []
page_number = 1

while True:
    # ê° í˜ì´ì§€ì˜ URL
    url = f'https://www.musinsa.com/ranking/best?period=now&age=ALL&mainCategory=&subCategory=&leafCategory=&price=&golf=false&kids=false&newProduct=false&exclusive=false&discount=false&soldOut=false&page={page_number}&viewType=small&priceMin=&priceMax='
    print(f"í˜„ì¬ {page_number} í˜ì´ì§€ ìˆ˜ì§‘ ì¤‘")
    # í˜ì´ì§€ ë‚´ìš©ì„ ê°€ì ¸ì˜´
#     response = requests.get(url)
#     soup = BeautifulSoup(response.text, 'html.parser')

    # í˜„ì¬ í˜ì´ì§€ì˜ ì œí’ˆ ì •ë³´ ìˆ˜ì§‘
    time.sleep(5)
    products_on_page = soup2.select('.li_box')
    for i in products_on_page:
        title = i.select_one('.item_title').text.strip()
        product_info = i.select_one('.list_info').text.strip()
        price = i.select_one('.price').text.strip().replace(" ", "")
        discount = i.select_one('.mu-icon-coupon > span').text if i.select_one('.mu-icon-coupon > span') is not None else ""
        review = i.select_one('p.point > span.count').text.strip() if i.select_one('p.point > span.count') is not None else ""   
        good = int(i.select_one("div.article_info > p.txt_cnt_like").text.replace(",", ""))
        
        product_list.append([title, product_info, price, discount, review, good])

    # ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™í•˜ëŠ” ë¡œì§
    if page_number == 100:
        break  # ë‹¤ìŒ í˜ì´ì§€ê°€ ì—†ìœ¼ë©´ ì¢…ë£Œ
    page_number += 1

# ìˆ˜ì§‘í•œ ì •ë³´ë¡œ DataFrame ìƒì„±
df = pd.DataFrame(product_list, columns=['Title', 'Product_Info', 'Price', 'Discount', 'Review', 'Good'])

# DataFrame ì¶œë ¥
print(df)
```
