```python
# url 물음표 전까지 가져오기
url = 'https://api.odcloud.kr/api/3074462/v1/uddi:7e73ac26-d0fa-4444-9c2e-81fdad4eff1b_201912061355'
```

```python
# decoding 키 가져오기
payload = dict(serviceKey="CYfhWwsG3T6CbKfbZn+oWe5Ks2hKiIksiycfndf9OYy2l2XlQL4s1y1L5kDlVqe0JzyiTTRcsSO8Np03ikBXXA==",
              page=1,
              perPage=100,
              returnType="xml")
```

```python
r = requests.get(url,params=payload)
```

```python
r.text
```

```python
crime_list = []
for i in json.loads(r.text)['data']:
    info = i
    crime_list.append(info)
```

```python
crime_list
```

```python
rows = []

# 주어진 JSON 데이터로부터 데이터프레임의 행 데이터 생성
for entry in crime_list:
    crime_category = f"{entry['범죄대분류']} - {entry['범죄중분류']}"
    for region, value in entry.items():
        if region not in ['범죄대분류', '범죄중분류']:
            rows.append({'지역명': region, '범죄유형': crime_category, '값': value})
```

```python
df = pd.DataFrame(rows)
```

```python
data = df.set_index("지역명")
```
