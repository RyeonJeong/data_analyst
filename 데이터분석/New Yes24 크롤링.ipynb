```python
import requests
```

```python
url = 'https://www.yes24.com/24/category/bestseller'
```

```python
payload = dict(CategoryNumber='001',sumgb='09',PageNumber=1)
payload
```

```python
r = requests.get(url, params=payload)
print(r.url)
print(r.status_code)
```

```python
response = r.text
```

```python
response
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(response, 'lxml') # 기본은 html.parser 이지만 빠른 속도를 위해서 lxml사용
soup
```

```python
best_list = soup.select('#bestContentsWrap')
```

```python
def refine(x):
    x = x.strip("-, >,→,\n")
    x = x.replace(",","")
    
    return x

# 변수이름.replace("바꿀내용", "바뀐내용")
# 변수이름.strip("없앨거1, 없앨거2")
```

```python
# refine 함수를 사용하여 텍스트를 정제하고 선택한 요소의 텍스트를 출력
all_list = soup.select_one('.bSGoodsSecArea')

if all_list:
    text = refine(all_list.text)
    print(text)
else:
    print("해당 요소를 찾을 수 없습니다.")
```
