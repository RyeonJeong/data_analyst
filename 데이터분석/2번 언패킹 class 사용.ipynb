```python
import json
import pandas as pd
```

```python
with open(r"C:\bigdata\data\전국치킨집리스트.json", "r") as file:
    data = file.read()
```

```python
# json 데이터를 클래스로 언패킹할 클래스 정의
```

```python
class Chicken:
    def __init__(self,data): # 자료로딩 생성자
        self.data = data
        
    def make_json(self,name): # json 자료 읽어오기
        with open(name, "r") as file:
            self.data = json.load(file)
    def unpack_json(self): # 자료 언패킹
        shop_info = []
        
        for shops in self.data:
            for sp_info in shops['body']['items']:
                temp = list(sp_info.values())
                shop_info.append(temp)
        return shop_info
    
    def make_csv(self,csv_name): # 자료 저장
        result = pd.DataFrame()
        for item in self.data:
            temp = pd.DataFrame(item['body']['items'])
            result = pd.concat([result, temp])
            
        result.to_csv(csv_name, encoding='utf-8')
```

```python
test = Chicken(None)
```

```python
test.make_json(r"C:\bigdata\data\전국치킨집리스트.json")
```
